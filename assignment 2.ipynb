```python
from html.parser import HTMLParser
class HeadingParser(HTMLParser):
    def handle_starttag(self, tags, att):
        if tags == 'data':
            for attributes in att:
                if attr[0] == 'data':
                    print(attr[1],end="")
    def handle_data(self, data):
        print(data,end="")
infile = open('w3c.html')
content = infile.read()
infile.close()
hp = HeadingParser()
hp.feed(content)
```
